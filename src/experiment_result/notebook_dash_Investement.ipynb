## Shapley values per Expert

```python
#### join to_shapley_by_source_table() with bundle.to_choquet_table().query("stage=='pre'") by expert and company
merged_df = pd.merge(
    bundle.to_shapley_by_source_table(),
    bundle.to_choquet_table().query("stage=='pre'"),
    left_on=['source', 'company'], right_on=['expert_id', 'company'], how='inner')

# 3. Select and reorder the required final columns
bf_consensus_shapley_choquet_by_expert = merged_df[['expert_id', 'company', 'shapley', 'score']]
bf_df = bf_consensus_shapley_choquet_by_expert.copy()


# Create the 'shapley_rank' column
bf_df['shapley_rank'] = bf_df.groupby('expert_id')['shapley'].rank(
    method='dense', # 'dense' does not skip ranks on ties
    ascending=False # Rank 1 goes to the highest value
).astype(int)

# Create the 'score_rank' column
bf_df['choquet_score_rank'] = bf_df.groupby('expert_id')['score'].rank(
    method='dense', # 'dense' does not skip ranks on ties
    ascending=False # Rank 1 goes to the highest value
).astype(int)

bf_df_sorted = bf_df.sort_values(by=['expert_id', 'shapley_rank'],ascending=[True, True] )
bf_df_sorted.head()
```

```python
chart = plot_topk_shapley_bars_unified(bf_df, top_k=10, shapley_col='shapley', expert_col="expert_id", columns=3, as_percent=True)
chart  # display in notebook
```

## final shapely values and interaction table

```python
af_df = bundle.to_shapley_choquet_table()
# Create the 'shapley_rank' column
af_df['final_shapley_rank'] = af_df['final_shapley'].rank(
    method='dense', # 'dense' does not skip ranks on ties
    ascending=False # Rank 1 goes to the highest value
).astype(int)

# Create the 'score_rank' column
af_df['final_choquet_rank'] = af_df['final_choquet'].rank(
    method='dense', # 'dense' does not skip ranks on ties
    ascending=False # Rank 1 goes to the highest value
).astype(int)

af_df_sorted = af_df.sort_values(by=['final_shapley_rank'],ascending=[True] )

top_n_companies = af_df_sorted['company'].head(10).tolist()

print(top_n_companies)
af_df_sorted

```
